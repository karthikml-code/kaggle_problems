```python
#finding all the null value columns so that we can fill it with mean value
na_col = train_df.columns[train_df.isna().any()].to_list()
```

```python
na_obj_col = [col for col in train_df[na_col].columns if train_df[col].dtypes == 'object']
na_flt_col = [col for col in train_df[na_col].columns if train_df[col].dtypes == 'float64']
```

```python
mode = train_df.mode()
#we can use the manual method for median or we can use the imputer method
#median = train_df[na_flt_col].median()
```

```python
mode
```

```python
#we used simple imputer function here to fill the na values with mean only for float columns
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(train_df[na_flt_col])
train_df[na_flt_col] = imputer.transform(train_df[na_flt_col])
```

```python
# we used manual fixing for the object columns using the mode functionality
train_df[na_obj_col] = train_df[na_obj_col].fillna(train_df.mode().iloc[0])
```

```python
#now we have 0 null values
train_df.isna().sum()
```

```python
#now we are going to encode the object columns with label encoder
column_objects
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
train_df[column_objects] = train_df[column_objects].apply(labelencoder.fit_transform)
```

```python
#data after we have encoded them.
train_df
```

```python
#all we have done is data preprocessing
#now we are going to split them into X and y
# here y is the loan status and the remaining columns are features. So the features will be in X
X = train_df.iloc[:,:-1].values
y = train_df.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

```python
#now we have split the data into X and y, we can go ahead and split them using train test split and train the models
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state=0)
```

```python
#we are doing feature scaling here. this increases models performance and accuracy
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
#this is done to print all the parameters of the classifier
from sklearn import set_config
set_config(print_changed_only=False)
```

```python
# we are using Kernel svm for this classification problem
from sklearn.svm import SVC
classifier = SVC(kernel='rbf',random_state=0)
classifier.fit(X_train,y_train)
```

```python
#we are going to predict the values for the X_test data
y_pred_svm = classifier.predict(X_test)
```

```python
#print the predicted and original values below
print(np.concatenate((y_pred_svm.reshape(len(y_pred_svm),1),y_test.reshape(len(y_test),1)),1))
```

```python
#calculation the accuracy scores, confusion matrix and classification report
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
confusion_mat_ksvm = confusion_matrix(y_test,y_pred_svm)
acc_score_ksvm = accuracy_score(y_test,y_pred_svm)
class_report_ksvm = classification_report(y_test,y_pred_svm)
```

```python
print('Confusion Matrix')
print(f"{confusion_mat_ksvm}\n")
print(f'Accuracy score: {acc_score_ksvm}\n')
print('Classification report:')
print(class_report_ksvm)
```

```python
#lets try out Randomforest classifier on the same dataset to get the accuracy score
#random state is not needed. you can use it if you want to see the same results as mine
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=2500,criterion='entropy',random_state=0)
rfc.fit(X_train,y_train)
```

```python
y_pred_rbc = rfc.predict(X_test)
print(np.concatenate((y_pred_rbc.reshape(len(y_pred_rbc),1),y_test.reshape(len(y_test),1)),1))
```

```python
confusion_mat_rfc = confusion_matrix(y_test,y_pred_rbc)
acc_score_rfc = accuracy_score(y_test,y_pred_rbc)
class_report_rfc = classification_report(y_test,y_pred_rbc)
```

```python
print('Confusion Matrix')
print(f"{confusion_mat_rfc}\n")
print(f'Accuracy score: {acc_score_rfc}\n')
print('Classification report:')
print(class_report_rfc)
```
